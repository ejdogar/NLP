```python
from keras.layers import Dense,SimpleRNN
from keras.models import Sequential
```

```python
model = Sequential()

model.add(SimpleRNN(3, input_shape=(4,5)))
model.add(Dense(1, activation="sigmoid"))

model.summary()
```

```python
learnable_parameters = 0
```

```python
# weights between input layer to hidden layer
print(model.get_weights()[0].shape)
print(model.get_weights()[0])

learnable_parameters = model.get_weights()[0].shape[0] * model.get_weights()[0].shape[1]
print(f"Learnable parameters are: {learnable_parameters}")
```

```python
# weights at the hidden layer of 3 nodes
print(model.get_weights()[1].shape)
print(model.get_weights()[1])

learnable_parameters += model.get_weights()[1].shape[0] * model.get_weights()[1].shape[1]
print(f"Learnable parameters are: {learnable_parameters}")
```

```python
# bias weights of each node at hidden layer
print(model.get_weights()[2].shape)
print(model.get_weights()[2])

learnable_parameters += model.get_weights()[2].shape[0]
print(f"Learnable parameters are: {learnable_parameters}")
```

```python
# weights between the hidden layer and output layer
print(model.get_weights()[3].shape)
print(model.get_weights()[3])

learnable_parameters += model.get_weights()[3].shape[0] * model.get_weights()[3].shape[1]
print(f"Learnable parameters are: {learnable_parameters}")
```

```python
# bias weight of the output node
print(model.get_weights()[4].shape)
print(model.get_weights()[4])

learnable_parameters += model.get_weights()[4].shape[0]
print(f"Learnable parameters are: {learnable_parameters}")
```
