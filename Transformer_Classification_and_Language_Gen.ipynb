```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
#importing a pre-trained model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
# tokenizing our model, "distilbert-base-uncased-finetuned-sst-2-english" this is tokenizer model name
tokenizer = AutoTokenizer.from_pretrained(model_name)
#loading our classification model
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
#text that we want to classify
text = "I really like this movie"
#tokenizzing our text; return_tensors = "pt" : this will return us pytorch tensors
input_ids = tokenizer(text, truncation = True, return_tensors="pt")["input_ids"] # return: tensor([[ 101, 1045, 2428, 2066, 2023, 3185,  102]])

output = model(input_ids, output_hidden_states = False, output_attentions = False, return_dict = True)

print(f"these are the outputs: --------------------------------- \n {output} \n ---------------------------- output ends here")

#convertion of output into probabilities
prob = torch.softmax(output["logits"][0], -1).tolist()

label_name = model.config.id2label.values()

predict = {name: round(float(pred) * 100) for pred, name in zip(prob, label_name)}

print(predict)
```


# Language Generation Model

A plain decoder based language model used for language generation.

```python
from transformers import AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

prompt = "it is the best"

# tokenizing our inputs
inputs = tokenizer(prompt, return_tensors = "pt")

outputs = model.generate(**inputs, max_length=100)

# as out output is in the form of tensors, so we will decode it using tokenizer

outputs_decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)

print(outputs_decoded)
```
